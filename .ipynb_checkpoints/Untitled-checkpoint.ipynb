```python
# === Setup + Step 0: mock data for multiple sheets ===
import numpy as np
from typing import Callable, Sequence

Array = np.ndarray
np.set_printoptions(precision=6, suppress=True)

def generate_mock_data(
    funcs: Sequence[Callable[[Array], Array]],
    X: Array,
    noise_std: float = 0.0,
    sort_values: bool = True,
    seed: int | None = 0,
):
    """
    Generate mock data for a list of input sheet functions f_i.
    Each f_i must accept X of shape (n, d) and return shape (n,).
    Returns (X_(n,d), Y_raw_(n,m), Y_sorted_(n,m)).
    """
    X = np.asarray(X)
    if X.ndim == 1:
        X = X[:, None]
    n, d = X.shape
    m = len(funcs)

    Y = np.zeros((n, m), dtype=float)
    for j, f in enumerate(funcs):
        vals = f(X)
        if vals.shape != (n,):
            raise ValueError(f"Function {j} must return shape (n,), got {vals.shape}")
        Y[:, j] = vals

    if noise_std > 0:
        rng = np.random.default_rng(seed)
        Y = Y + noise_std * rng.standard_normal(Y.shape)

    Y_sorted = np.sort(Y, axis=1) if sort_values else Y.copy()
    return X, Y, Y_sorted


# --- demo functions: three crossing 1D sheets on [-1, 1] ---
def f1(X):
    x = X[:, 0]
    return 0.6 * x + 0.10 * np.cos(4 * np.pi * x)

def f2(X):
    x = X[:, 0]
    return -0.5 * x + 0.08 * np.sin(5 * np.pi * x)

def f3(X):
    x = X[:, 0]
    return 0.2 * np.sin(3 * np.pi * x)

funcs = [f1, f2, f3]

# --- build training grid and generate mock observations ---
x_train = np.linspace(-1, 1, 161)        # tweak density if you like
Xtr = x_train[:, None]
Xtr, Y_raw, Y_sorted = generate_mock_data(funcs, Xtr, noise_std=0.0, sort_values=True)

print("Shapes -> Xtr:", Xtr.shape, " Y_raw:", Y_raw.shape, " Y_sorted:", Y_sorted.shape)
print("First 3 rows of Y_raw:\n", Y_raw[:3])
print("First 3 rows of Y_sorted:\n", Y_sorted[:3])
```

```python
# === Step 1: compute elementary symmetric polynomials (ESPs) ===

def values_to_esps(Y: np.ndarray) -> np.ndarray:
    """
    For each row y = (y1,...,ym), form the monic polynomial
        p(t) = ∏ (t - y_i) = t^m - s1 t^{m-1} + s2 t^{m-2} - ... + (-1)^m sm
    and return s = [s1,...,sm].
    We extract s_k via numpy.poly's coefficients.
    """
    Y = np.asarray(Y)
    if Y.ndim != 2:
        raise ValueError("Y must be 2D (n, m)")
    n, m = Y.shape
    S = np.zeros((n, m), dtype=float)
    for i in range(n):
        coeff = np.poly(Y[i, :])  # [1, -s1, s2, ..., (-1)^m sm]
        for k in range(1, m + 1):
            S[i, k - 1] = (-1) ** k * coeff[k]
    return S

# Use Y_raw (order doesn't matter)
S = values_to_esps(Y_raw)
print("S shape:", S.shape)
print("First 3 rows of S:\n", S[:3])

# (Sanity check for the very first row: with y=[-0.5, 0.5, 0] expect s=[0, -0.25, 0])
```

```python
# === Step 2: learn smooth ESP surfaces (total-degree polynomial regression) ===
import numpy as np
from dataclasses import dataclass
from typing import List, Tuple

def _total_degree_powers(d: int, degree: int) -> List[Tuple[int, ...]]:
    # All nonnegative integer tuples e=(e1..ed) with sum(e) <= degree.
    powers: List[Tuple[int, ...]] = []

    def rec(prefix: Tuple[int, ...], remaining_vars: int, budget: int):
        if remaining_vars == 0:
            powers.append(prefix)
            return
        if remaining_vars == 1:
            powers.append(prefix + (budget,))
            return
        for a in range(budget + 1):
            rec(prefix + (a,), remaining_vars - 1, budget - a)

    for tot in range(degree + 1):
        rec((), d, tot)
    return powers

def _build_design_matrix(X: np.ndarray, powers: List[Tuple[int, ...]]) -> np.ndarray:
    X = np.asarray(X)
    if X.ndim == 1:
        X = X[:, None]
    n, d = X.shape
    Phi = np.ones((n, len(powers)), dtype=float)
    for j, exp in enumerate(powers):
        col = np.ones(n, dtype=float)
        for k in range(d):
            if exp[k] != 0:
                col *= X[:, k] ** exp[k]
        Phi[:, j] = col
    return Phi

@dataclass
class PolySurface:
    powers: List[Tuple[int, ...]]
    coef: np.ndarray               # (n_terms,)
    x_min: np.ndarray              # for scaling to [-1,1]
    x_max: np.ndarray

    def _scale(self, X: np.ndarray) -> np.ndarray:
        X = np.asarray(X)
        if X.ndim == 1:
            X = X[:, None]
        eps = 1e-12
        denom = np.maximum(self.x_max - self.x_min, eps)
        Z = -1 + 2 * (X - self.x_min) / denom
        return Z

    def predict(self, X: np.ndarray) -> np.ndarray:
        Z = self._scale(X)
        Phi = _build_design_matrix(Z, self.powers)
        return Phi @ self.coef

class PolyFitter:
    def __init__(self, degree: int = 8, ridge: float = 1e-10):
        self.degree = degree
        self.ridge = ridge

    def fit(self, X: np.ndarray, y: np.ndarray) -> PolySurface:
        X = np.asarray(X)
        if X.ndim == 1:
            X = X[:, None]
        x_min = X.min(axis=0)
        x_max = X.max(axis=0)

        # Scale features to [-1, 1] for conditioning
        eps = 1e-12
        denom = np.maximum(x_max - x_min, eps)
        Z = -1 + 2 * (X - x_min) / denom

        powers = _total_degree_powers(Z.shape[1], self.degree)
        Phi = _build_design_matrix(Z, powers)
        if self.ridge > 0:
            A = Phi.T @ Phi + self.ridge * np.eye(Phi.shape[1])
            b = Phi.T @ y
            coef = np.linalg.solve(A, b)
        else:
            coef, *_ = np.linalg.lstsq(Phi, y, rcond=None)
        return PolySurface(powers=powers, coef=coef, x_min=x_min, x_max=x_max)

def learn_esp_surfaces(X: np.ndarray, S: np.ndarray, degree: int = 8, ridge: float = 1e-10):
    """Fit one polynomial surface per ESP column in S."""
    X = np.asarray(X)
    if X.ndim == 1:
        X = X[:, None]
    m = S.shape[1]
    fitter = PolyFitter(degree=degree, ridge=ridge)
    models = []
    for k in range(m):
        models.append(fitter.fit(X, S[:, k]))
    return models

def predict_esps(models, Xq: np.ndarray) -> np.ndarray:
    Xq = np.asarray(Xq)
    if Xq.ndim == 1:
        Xq = Xq[:, None]
    nq = Xq.shape[0]
    m = len(models)
    S_hat = np.zeros((nq, m), dtype=float)
    for k, model in enumerate(models):
        S_hat[:, k] = model.predict(Xq)
    return S_hat

# --- fit ESP models on your data from Steps 0–1 ---
models = learn_esp_surfaces(Xtr, S, degree=8, ridge=1e-10)
print("Fitted", len(models), "ESP surfaces.")

# quick sanity check (training fit)
S_fit = predict_esps(models, Xtr)
rmse_per_esp = np.sqrt(np.mean((S_fit - S) ** 2, axis=0))
print("Training RMSE per ESP:", rmse_per_esp)
```

```python
# === Step 3: build Frobenius companion matrices from the learned ESPs ===
import numpy as np

def companion_from_esps(s: np.ndarray) -> np.ndarray:
    """
    Given s = [s1,...,sm], form the Frobenius companion matrix of
      p(t) = t^m - s1 t^{m-1} + s2 t^{m-2} - ... + (-1)^m sm.
    """
    s = np.asarray(s, dtype=float).ravel()
    m = s.size
    # rewrite p(t) = t^m + a_{m-1} t^{m-1} + ... + a_0
    a = np.array([(-1)**(m - k) * s[m - 1 - k] for k in range(m)], dtype=float)
    C = np.zeros((m, m), dtype=float)
    C[1:, :-1] = np.eye(m - 1)   # subdiagonal ones
    C[:, -1] = -a                # last column
    return C

# choose query points and predict ESPs there
x_query = np.linspace(-1, 1, 401)
Xq = x_query[:, None]
S_hat = predict_esps(models, Xq)   # uses the 'models' from Step 2

# (optional) inspect one example companion matrix at the midpoint (x ≈ 0)
C_mid = companion_from_esps(S_hat[len(x_query)//2])
print("S_hat shape:", S_hat.shape, " (rows = points, cols = [s1,s2,s3])")
print("Example companion matrix at x≈0:\n", C_mid)
```

```python
# === Step 4: reconstruct original sheet values via eigenvalues ===
import numpy as np

def reconstruct_values_from_esps(S_hat: np.ndarray, enforce_real: bool = True) -> np.ndarray:
    nq, m = S_hat.shape
    Y_rec = np.zeros((nq, m), dtype=float)
    for i in range(nq):
        C = companion_from_esps(S_hat[i])  # uses function from Step 3
        eig = np.linalg.eigvals(C)
        if enforce_real and np.max(np.abs(eig.imag)) <= 1e-8:
            eig = eig.real
        else:
            # keep the real part if tiny imaginary noise remains
            eig = eig.real
        Y_rec[i] = eig
    return Y_rec

# Reconstruct on your query grid
Y_rec = reconstruct_values_from_esps(S_hat, enforce_real=True)

# Evaluate against ground truth
Y_true_q = np.column_stack([f(Xq) for f in funcs])  # funcs, Xq defined earlier
Y_true_sorted = np.sort(Y_true_q, axis=1)
Y_rec_sorted  = np.sort(Y_rec, axis=1)

mae = np.mean(np.abs(Y_rec_sorted - Y_true_sorted))
print(f"Mean Absolute Error (sorted): {mae:.3e}")

# A few checkpoints
for xi in [-1.0, -0.5, 0.0, 0.5, 1.0]:
    idx = np.argmin(np.abs(x_query - xi))
    print(f"x={x_query[idx]: .3f} | true={Y_true_sorted[idx]} | recon={Y_rec_sorted[idx]}")
```

```python
# Visualization 1: true vs reconstructed (sorted) and absolute error per sheet
import numpy as np
import matplotlib.pyplot as plt

# assumes: x_query, Y_true_sorted, Y_rec_sorted already exist from Step 4
per_sheet_mae  = np.mean(np.abs(Y_rec_sorted - Y_true_sorted), axis=0)
per_sheet_rmse = np.sqrt(np.mean((Y_rec_sorted - Y_true_sorted)**2, axis=0))
print("Per-sheet MAE :", per_sheet_mae)
print("Per-sheet RMSE:", per_sheet_rmse)

# --- Figure A: true vs reconstructed (sorted) ---
plt.figure(figsize=(8, 4.5))
for j in range(Y_true_sorted.shape[1]):
    plt.plot(x_query, Y_true_sorted[:, j], label=f"true sheet {j+1}", linewidth=2)
for j in range(Y_rec_sorted.shape[1]):
    plt.plot(x_query, Y_rec_sorted[:, j], linestyle="--", label=f"recon sheet {j+1}")
plt.xlabel("x")
plt.ylabel("value")
plt.title("Sorted sheets: true vs reconstructed")
plt.legend()
plt.show()

# --- Figure B: absolute error per sheet ---
abs_err = np.abs(Y_rec_sorted - Y_true_sorted)
plt.figure(figsize=(8, 4.5))
for j in range(abs_err.shape[1]):
    plt.plot(x_query, abs_err[:, j], label=f"|error| sheet {j+1}")
plt.xlabel("x")
plt.ylabel("absolute error")
plt.title("Pointwise absolute error (sorted sheets)")
plt.legend()
plt.show()
```
